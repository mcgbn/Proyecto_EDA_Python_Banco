## Homogeneizamos texto en columnas categoricas

`````python
#para poner todas las columnas con string en minuscula, reemplazar a barra baja los espacios y separaciones por puntos

for col in datos_bancarios.select_dtypes(include='O').columns:
    datos_bancarios[col]= datos_bancarios[col].str.lower()
`````

`````python
#para poner todas las columnas con string con espacio en lugar de punto
for col in datos_bancarios.select_dtypes(include='object').columns:
    datos_bancarios[col]= datos_bancarios[col].str.replace(".", "_")

#equivalente en excel de hacer operacion con una columna, se elimina la anterior y se renombra con el nombre de la columna eliminada.
`````

`````python
#comprobamos que esté en minúsculas y separado por espacios
datos_bancarios.sample(1)
`````

## Pasar datos a fecha

`````python
#panda tiene instalado datetime, pueden utilizarse los metodos directamente
`````

`````python
datos_bancarios.sample(2)
`````

`````python
#al estar en español, necesitamos cambiar el mes a inglés. Uso de diccionarios:

meses_es = {
    "enero": "01",
    "febrero": "02",
    "marzo": "03",
    "abril": "04",
    "mayo": "05",
    "junio": "06",
    "julio": "07",
    "agosto": "08",
    "septiembre": "09",
    "setiembre": "09",  # variante válida en español
    "octubre": "10",
    "noviembre": "11",
    "diciembre": "12"
}
`````

`````python
datos_bancarios.replace({'date':meses_es}, regex=True, inplace=True)
datos_bancarios.sample(5)
`````

`````python
#una vez cambiado de español a imglés, lanzar para convertir a fecha
datos_bancarios['date'] = pd.to_datetime(datos_bancarios['date'], format='%d-%m-%Y')
`````

`````python
datos_bancarios.info()
`````

## Cambiar comas por puntos

`````python
for col in datos_bancarios.select_dtypes(include='O').columns:
    datos_bancarios[col]= datos_bancarios[col].str.replace(',','.')
    try:
        datos_bancarios[col]=datos_bancarios[col].astype(float)
    except:
        pass
`````
